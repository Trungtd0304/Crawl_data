```python
import undetected_chromedriver as uc
import time 
import pandas as pd
from datetime import date
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

driver = uc.Chrome(version_main=119)
today = date.today()
driver.get('https://shopee.vn/Thi%E1%BA%BFt-B%E1%BB%8B-%C4%90i%E1%BB%87n-T%E1%BB%AD-cat.11036132')
time.sleep(3)
# Vòng lập for lấy thông tin số trang
for i in range(10):  # Thay đổi số lần scroll tùy ý
    # Sử dụng phím mũi tên xuống để scroll down
    driver.find_element("css selector", 'body').send_keys(Keys.ARROW_DOWN)
total_span = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, "div.shopee-mini-page-controller__state > span.shopee-mini-page-controller__total"))
)
# Lấy nội dung của thẻ span
total = total_span.text

list_lst =[]
for x in range(0,int(total)):
    driver.get(f'https://shopee.vn/Thi%E1%BA%BFt-B%E1%BB%8B-%C4%90i%E1%BB%87n-T%E1%BB%AD-cat.11036132?page={x}')
    time.sleep(10)
    # Lặp lại việc scroll down nhiều lần
    for i in range(150):  # Thay đổi số lần scroll tùy ý
        # Sử dụng phím mũi tên xuống để scroll down
        driver.find_element("css selector", 'body').send_keys(Keys.ARROW_DOWN)
        # Chờ một khoảng thời gian ngắn giữa các lần scroll (để tránh tốc độ quá nhanh)
        time.sleep(0.1)
    # get list element in page
    lst = [element.get_attribute("href") for element in driver.find_elements(By.CSS_SELECTOR, "li[class*=shopee-search-item-result__item] a")]
    list_lst.append(lst)
    time.sleep(5)

data_products, data_shop = [],[]
for i in range(0,1):
    cnt = 0
    for itm in list_lst[i]: 
        if itm == None or itm == "":
            continue

        try:
            driver.get(itm) # go to each detail page
            time.sleep(3)
        except:
            print(itm)

        for i in range(100):  # Thay đổi số lần scroll tùy ý
            # Sử dụng phím mũi tên xuống để scroll down
            driver.find_element("css selector", 'body').send_keys(Keys.ARROW_DOWN)
        # Thông tin đơn hàng
        title = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=_44qnta] > span"))).text
        try:
            danh_gia = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "div._1k47d8:not(._046PXf)"))).text
        except TimeoutException:
            danh_gia = None
        try:
            da_ban = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "div.e9sAa2"))).text
        except TimeoutException:
            da_ban = None
        gia_ban = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.pqTWkA"))).text
        so_luong = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "section._6lioXX > div:nth-child(2) > div:nth-child(2)"))).text
        try:
            luot_like = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=IYjGwk] > [class*=Ne7dEf]"))).text
        except TimeoutException:
            luot_like = None
        tinh_gui = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.dR8kXc:nth-last-child(1) >div"))).text
        
        # Thông tin sell
        shop_name = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=oAVg4E] > [class*=VlDReK]"))).text
        link_shop =driver.find_element(By.CSS_SELECTOR, "[class*=jwlMoy] > a").get_attribute("href")
        danh_gia_shop = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(1) > span"))).text
        ty_le_phan_hoi = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(2) > span"))).text
        mo_shop = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(3) > span"))).text
        sp_shop = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(4) > span"))).text
        thoi_gian_phan_hoi = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(5) > span"))).text
        luot_theo_doi_shop = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(6) > span"))).text
            
        try:
            shop_yeu_thich = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "[class*=NP8R3A] > [class*=NOygQS]"))).text
            star = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "div._1k47d8._046PXf"))).text
            danh_gia = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "div._1k47d8:not(._046PXf)"))).text
        except TimeoutException:
            shop_yeu_thich = None
            star = None
            danh_gia = None
        # Thông tin đơn hàng
        time.sleep(1)
        try:
            sp_yeu_thich = driver.find_element(By.CSS_SELECTOR, "[class*=NOygQS]").text
            time.sleep(1)
            giam_gia = driver.find_element(By.CSS_SELECTOR, "div._0voski").text
        except NoSuchElementException:
            # Xử lý khi không tìm thấy phần tử
            sp_yeu_thich = None
            giam_gia = None

        cnt += 1
        # df chứa thông tin sản phẩm
        data_products.append({
            'Date crawl':today,
            'Tên sản phẩm': title,
            'Sản phẩm yêu thích': sp_yeu_thich,
            'Điểm đánh giá': star,
            'Số lượt đánh giá': danh_gia,
            'Sản lượng đã bán': da_ban,
            'Giá sản phẩm hiện tại': gia_ban,
            'Giảm giá': giam_gia,
            'Tồn kho': so_luong,
            'Lượt thích': luot_like,
            'Tên shop': shop_name,
            'Tỉnh gửi hàng': tinh_gui,
            'Link sản phẩm':itm
        })
        # df chứa thông tin shop
        data_shop.append({
            'Date crawl':today,
            'Tên shop': shop_name,
            'Tỉnh gửi hàng': tinh_gui,
            'Thời gian tham gia sàn': mo_shop,
            'Shop yêu thích': shop_yeu_thich,
            'Đánh giá shop': danh_gia_shop,
            'Tỷ lệ phản hồi': ty_le_phan_hoi,
            'Thời giản phản hồi': thoi_gian_phan_hoi,
            'Số lượng sản phẩm bán': sp_shop,
            'Số lượng người theo dõi': luot_theo_doi_shop,
            'Link shop': link_shop
        })

df_products = pd.DataFrame(data_products)
df_shop = pd.DataFrame(data_shop)

driver.close()
```
