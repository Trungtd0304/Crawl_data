```python
import undetected_chromedriver as uc
import time 
import pandas as pd
from selenium.webdriver.common.by import By 
from datetime import date
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException

driver = uc.Chrome(version_main=119)
today = date.today()
driver.get('https://shopee.vn/Thi%E1%BA%BFt-B%E1%BB%8B-%C4%90i%E1%BB%87n-T%E1%BB%AD-cat.11036132?page=0')
time.sleep(20)
# Lặp lại việc scroll down nhiều lần
for i in range(200):  # Thay đổi số lần scroll tùy ý
    # Sử dụng phím mũi tên xuống để scroll down
    driver.find_element("css selector", 'body').send_keys(Keys.ARROW_DOWN)
    # Chờ một khoảng thời gian ngắn giữa các lần scroll (để tránh tốc độ quá nhanh)
    time.sleep(0.1)

# get list element in page
lst = [element.get_attribute("href") for element in driver.find_elements(By.CSS_SELECTOR, "li[class*=shopee-search-item-result__item] a")]


data_products, data_shop = [],[]
cnt = 0
for itm in lst: 
    if itm == None or itm == "":
        continue

    try:
        driver.get(itm) # go to each detail page
        time.sleep(3)
    except:
        print(itm)

    # Thông tin đơn hàng
    try:
        sp_yeu_thich = driver.find_element(By.CSS_SELECTOR, "[class*=NOygQS]").text
        title = driver.find_element(By.CSS_SELECTOR, "[class*=_44qnta] > span").text
        star = driver.find_element(By.CSS_SELECTOR, "div._1k47d8._046PXf").text
        danh_gia = driver.find_element(By.CSS_SELECTOR, "div._1k47d8:not(._046PXf)").text
        da_ban = driver.find_element(By.CSS_SELECTOR, "div.e9sAa2").text
        gia_ban = driver.find_element(By.CSS_SELECTOR, "div.pqTWkA").text
        giam_gia = driver.find_element(By.CSS_SELECTOR, "div._0voski").text
        so_luong = driver.find_element(By.CSS_SELECTOR, "section._6lioXX > div:nth-child(2) > div:nth-child(2)").text
        luot_like = driver.find_element(By.CSS_SELECTOR, "[class*=IYjGwk] > [class*=Ne7dEf]").text
        tinh_gui = driver.find_element(By.CSS_SELECTOR, "div.dR8kXc:nth-last-child(1) >div").text
    except NoSuchElementException:
        # Xử lý khi không tìm thấy phần tử
        sp_yeu_thich = None
        title = None
        star = None
        danh_gia = None
        da_ban = None
        gia_ban = None
        giam_gia = None
        so_luong = None
        luot_like = None
        tinh_gui = None
    
    # Thông tin sell
    try:
        shop_name = driver.find_element(By.CSS_SELECTOR, "[class*=oAVg4E] > [class*=VlDReK]").text
        link_shop = driver.find_element(By.CSS_SELECTOR, "[class*=jwlMoy] > a").get_attribute("href")
        shop_yeu_thich = driver.find_element(By.CSS_SELECTOR, "[class*=NP8R3A] > [class*=NOygQS]").text
        danh_gia_shop = driver.find_element(By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(1) > span").text
        ty_le_phan_hoi = driver.find_element(By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(2) > span").text
        mo_shop = driver.find_element(By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(3) > span").text
        sp_shop = driver.find_element(By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(4) > span").text
        thoi_gian_phan_hoi = driver.find_element(By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(5) > span").text
        luot_theo_doi_shop = driver.find_element(By.CSS_SELECTOR, "[class*=R7Q8ES]:nth-child(6) > span").text
    except NoSuchElementException:
        shop_name = None
        link_shop = None
        shop_yeu_thich = None
        danh_gia_shop = None
        ty_le_phan_hoi = None
        mo_shop = None
        sp_shop = None
        thoi_gian_phan_hoi = None
        luot_theo_doi_shop = None

    cnt += 1
    # df chứa thông tin sản phẩm
    data_products.append({
        'Tên sản phẩm': title,
        'Sản phẩm yêu thích': sp_yeu_thich,
        'Điểm đánh giá': star,
        'Số lượt đánh giá': danh_gia,
        'Sản lượng đã bán': da_ban,
        'Giá sản phẩm hiện tại': gia_ban,
        'Giảm giá': giam_gia,
        'Tồn kho': so_luong,
        'Lượt thích': luot_like,
        'Tên shop': shop_name,
        'Tỉnh gửi hàng': tinh_gui
    })
    # df chứa thông tin shop
    data_shop.append({
        'Tên shop': shop_name,
        'Tỉnh gửi hàng': tinh_gui,
        'Thời gian tham gia sàn': mo_shop,
        'Shop yêu thích': shop_yeu_thich,
        'Đánh giá shop': danh_gia_shop,
        'Tỷ lệ phản hồi': ty_le_phan_hoi,
        'Thời giản phản hồi': thoi_gian_phan_hoi,
        'Số lượng sản phẩm bán': sp_shop,
        'Số lượng người theo dõi': luot_theo_doi_shop,
        'Link shop': link_shop
    })
df_products = pd.DataFrame(data_products)
df_shop = pd.DataFrame(data_shop)

driver.close()
```
